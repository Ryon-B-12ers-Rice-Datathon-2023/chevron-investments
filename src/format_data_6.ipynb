```python
import pandas as pd
```

```python
df = pd.read_csv(r'../data/Investment_Data_Full.csv')

pivot_df = df.pivot_table(index='Year', columns=['StateCode','MSN'],

values=['Amount', 'TotalNumberofInvestments', 'TotalAmountofAssistance', 'CO2 Emissions (Mmt)'])
```

```python
pivot_df.to_csv(r'../data/15-20_lstm_assistance_msn.csv')
```

```python
included = []

for column in pivot_df.columns:
    if column[0] == 'Amount':
        continue
    if (column[0], column[1]) not in included:
        included.append((column[0], column[1]))
    else:
        pivot_df.drop(columns=column, inplace=True)
```
