```python
import pandas as pd
import numpy as np
```

```python
df_assist = pd.read_csv(r'../data/assistance.csv')
df_assist.columns = ['StateCode', 'TotalNumberofInvestments', 'TotalAmountofAssistance', 'Year']
cols = df_assist.columns.tolist()
cols = cols[-1:] + cols[:-1]
df_assist = df_assist[cols]
```

```python
df_assist
```

```python
state_codes = np.repeat(pd.unique(df_assist['StateCode']), 2)
variable_names = np.empty_like(state_codes)
variable_names[::2] = 'TotalAmountofAssistance'
variable_names[1::2] = 'TotalNumberofInvestments'
```

```python
state_codes = pd.unique(df_assist['StateCode'])
variable_names = np.empty_like(state_codes)

indices = pd.MultiIndex.from_arrays([
    state_codes, 
    variable_names
])

```

```python
indices = pd.MultiIndex.from_arrays([
    state_codes, 
    variable_names
])

final_df = pd.DataFrame(0, index=np.arange(12), columns=indices)
final_df['Year'] = pd.unique(df_assist['Year'])

cols = final_df.columns.tolist()
cols = cols[-1:] + cols[:-1]
final_df = final_df[cols]

```

```python
final_df
```

```python
for state_code, variable in final_df.columns[1:]:
    subset = df_assist[df_assist['StateCode'] == state_code]
    for index, row in subset.iterrows():
        idx = final_df.loc[final_df['Year'] == row['Year']].index[0]
        var_value = row[variable]
        final_df.at[idx, (state_code, variable)] = var_value
```

```python
final_df.to_csv(r'../data/12-year_lstm_assistance.csv')
```
